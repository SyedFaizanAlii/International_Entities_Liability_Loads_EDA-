```python
# Text Color
def PrintColor(text: str, color=Fore.CYAN, style=Style.BRIGHT):
    "Prints color outputs using colorama using a text F-string"
    print(style + color + text + Style.RESET_ALL)

# Text For Main Heading
def print_blue_large(text):
    PrintColor(text, Fore.RED + Style.BRIGHT)

# Main Heading
def print_boxed_blue_heading(text):
    length = len(text) + 4
    print(f"\n{Style.BRIGHT}{Fore.RED}{'='*length}{Style.RESET_ALL}")
    print(f"{Style.BRIGHT}{Fore.YELLOW}| {text} |{Style.RESET_ALL}")
    print(f"{Style.BRIGHT}{Fore.RED}{'='*length}{Style.RESET_ALL}")

# Function to Overview Data
def data_overview(df):

    # Display head of the training dataset nicely
    print_boxed_blue_heading("The Head Of The Dataset is:")
    display(HTML(df.head(5).to_html(index=False).replace('<table border="1" class="dataframe">', '<table style="border: 2px solid blue;">')))

    print('\n')

    # Shapes of Train and Test
    print_boxed_blue_heading("Shape Information:")
    PrintColor(f"The Shape Of The Data is {df.shape} || No of Rows is : {df.shape[0]} and Columns is {df.shape[1]}", Fore.CYAN)
    print('\n')

    # Info of Both Datasets
    print_boxed_blue_heading("Dataset Information:")
    PrintColor(f"\nThe Info Of The Dataset", Fore.CYAN)
    df.info()

    # Describe Both
    print_boxed_blue_heading("Numerical Summary:")
    PrintColor(f"\nThe Numerical Summary is", Fore.CYAN)
    display(df.describe().style.set_caption("Data Summary").set_table_styles([{'selector': 'caption', 'props': [('color', 'blue')]}]))

    # Null Values in Data
    print_boxed_blue_heading("Null Values:")
    PrintColor("\nNull Values in Data", Fore.CYAN)
    print(df.isnull().sum())
    print('\n')

    # Duplicates Values in Data
    print_boxed_blue_heading("Duplicate Values:")
    PrintColor("\nDuplicates Values in Data", Fore.CYAN)
    print(df.duplicated().sum())
```

```python

# Function to transform of big values to easy analysis the data 
print_boxed_blue_heading("Transformation Of Data :")
def transform_values(df, column, method='log'):
    if method == 'log':
        # Apply logarithmic transformation
        df[column] = np.log10(df[column])
        transformed_column_name = f'log10_{column}'
        df.rename(columns={column: transformed_column_name}, inplace=True)
        print(f"{Fore.CYAN}{Style.BRIGHT}The Data Transform Successfully{Style.RESET_ALL}")
    return df

# Apply transformations
df = transform_values(df, 'total liabilities (USD) in Trillion')

# # Plotting function
# def plot_transformed_data(df, column):
#     plt.figure(figsize=(10, 6))
#     sns.histplot(df[column], kde=True, color='skyblue')
#     plt.title(f'Histogram of {column}')
#     plt.xlabel(column)
#     plt.ylabel('Frequency')
#     plt.show()

#     plt.figure(figsize=(10, 6))
#     sns.boxplot(df[column], color='lightgreen')
#     plt.title(f'Boxplot of {column}')
#     plt.xlabel(column)
#     plt.show()


# # plot_transformed_data(df, 'log10_total liabilities (USD) in Trillion')
```

```python
continuous_vars = ['log10_total liabilities (USD) in Trillion', 'price (USD)']
def data_outliers(df, continuous_vars):
    # Outliers in Data
    print_boxed_blue_heading("Outliers in Continuous Variables:")
    PrintColor("\nThe Total Outliers In The Dataset are:", Fore.CYAN)
    Q1 = df[continuous_vars].quantile(0.25)
    Q3 = df[continuous_vars].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df[continuous_vars] < (Q1 - 1.5 * IQR)) | (df[continuous_vars] > (Q3 + 1.5 * IQR))).sum()
    outliers_df = pd.DataFrame(outliers, columns=['Outliers'])
    return outliers_df

# Example usage:
outliers_df = data_outliers(df, continuous_vars)
display(HTML(outliers_df.to_html().replace('<table border="1" class="dataframe">', '<table style="border: 2px solid blue;">')))
```

```python
def remove_outliers(df, continuous_vars):
    # df_clean = df.copy()
    print_boxed_blue_heading("Removal Of Outliers:")
    for var in continuous_vars:
        Q1 = df[var].quantile(0.25)
        Q3 = df[var].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[var] >= lower_bound) & (df[var] <= upper_bound)]
    print(f"{Fore.CYAN}{Style.BRIGHT}The Outliers Removed Successfully{Style.RESET_ALL}")
    return df

# Example usage:
df = remove_outliers(df, continuous_vars)
```

```python
# Find the top 10 most frequent company origins
top_10_origins = df['company origin'].value_counts().head(10).index

# Filter the dataset for the top 10 most frequent company origins
top_10_data = df[df['company origin'].isin(top_10_origins)]

# Define a color palette with different colors for each bar
colors = sns.color_palette("husl", len(top_10_origins))

# Plot the frequency of the top 10 most frequent company origins
plt.figure(figsize=(15, 8))
sns.countplot(y='company origin', data=top_10_data, order=top_10_data['company origin'].value_counts().index, palette=colors)
plt.title('Frequency of Top 10 Company Origins')
plt.xlabel('Count')
plt.ylabel('Company Origin')
plt.show()
```

```python
# Filter the dataset for the top 10 most frequent company origins
top_10_data = df[df['company origin'].isin(top_10_origins)]
# Comparing average total liabilities by company origin
avg_liabilities_by_origin = top_10_data.groupby('company origin')['log10_total liabilities (USD) in Trillion'].mean().sort_values()
liabilities_colors = sns.color_palette("husl", len(avg_liabilities_by_origin))
plt.figure(figsize=(15,8))
avg_liabilities_by_origin.plot(kind='barh', color=liabilities_colors)
plt.title('Average Total Liabilities (USD) in Trillion by Company Origin')
plt.xlabel('Average Total Liabilities (USD) in Trillion')
plt.ylabel('Country')
plt.show()

# Comparing average stock price by company origin
avg_price_by_origin = top_10_data.groupby('company origin')['price (USD)'].mean().sort_values()
price_colors = sns.color_palette("Paired", len(avg_price_by_origin))
plt.figure(figsize=(15, 8))
avg_price_by_origin.plot(kind='barh', color=price_colors)
plt.title('Average Stock Price (USD) by Company Origin')
plt.xlabel('Average Stock Price (USD)')
plt.ylabel('Country')
plt.show()
# Top 10 companies by total liabilities
top_10_liabilities = df.nlargest(10, 'log10_total liabilities (USD) in Trillion')

# Plotting the top 10 companies by total liabilities
plt.figure(figsize=(14, 8))
sns.barplot(
    x='log10_total liabilities (USD) in Trillion', 
    y='Company', 
    data=top_10_liabilities, 
    palette='viridis'
)
plt.title('Top 10 Companies by Total Liabilities (USD) in Trillion')
plt.xlabel('Total Liabilities (USD) in Trillion')
plt.ylabel('Company')
plt.show()
```
